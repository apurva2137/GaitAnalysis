```python
import cv2
import torch
import numpy as np
import time
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
import pydensecrf.densecrf as dcrf
from pydensecrf.utils import unary_from_softmax

# Load YOLOv8n-Seg model
model = YOLO("yolov8m-seg.pt")  # Medium-sized model for better segmentation
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Class index for 'person' in COCO dataset
PEOPLE_CLASS = 0

def apply_crf(img, mask):
    """
    Applies CRF (Conditional Random Fields) for mask refinement.
    """
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    mask = mask.astype(np.float32) / 255.0
    mask = np.stack([1 - mask, mask], axis=0)

    height, width = img.shape[:2]
    spatial_compat = max(height, width) // 50
    bilateral_compat = max(height, width) // 20

    d = dcrf.DenseCRF2D(width, height, 2)
    unary = unary_from_softmax(mask)
    d.setUnaryEnergy(unary)
    d.addPairwiseGaussian(sxy=10, compat=3)
    d.addPairwiseBilateral(sxy=10, srgb=5, rgbim=img, compat=10)


    Q = d.inference(5)
    refined_mask = np.argmax(Q, axis=0).reshape(img.shape[:2]) * 255
    return refined_mask.astype(np.uint8)

def make_seg_mask(img):
    """
    Generates a refined silhouette mask using YOLOv8 segmentation.
    """
    results = model(img)

    H, W, _ = img.shape
    mask = np.zeros((H, W), dtype=np.uint8)

    for result in results:
        for box, seg, cls in zip(result.boxes.xyxy, result.masks.xy, result.boxes.cls):
            if int(cls) == PEOPLE_CLASS:  # Only process 'person' class
                seg = np.array(seg, dtype=np.int32)
                cv2.fillPoly(mask, [seg], 255)

    return mask

#Change _name_ to __name__
if __name__ == '__main__':
    video_path = 'test_vid.mp4'  # Replace with your video
    video = cv2.VideoCapture(video_path)

    frame_skip = 10
    print(f"Using frame skip rate: {frame_skip}")

    prev_time = time.time()
    frame_count = 0

    while video.isOpened():
        success, img = video.read()
        if not success:
            break

        img = cv2.resize(img, (900, 1024))  # Upscale for better edge detection

        frame_count += 1
        if frame_count % frame_skip != 0:
            continue

        # Generate segmentation mask
        mask = make_seg_mask(img)

        # Refine using CRF
        mask = apply_crf(img, mask)

        mask = cv2.resize(mask, (900, 1024), interpolation=cv2.INTER_LINEAR)

        kernel = np.ones((3, 3), np.uint8)
        mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)  # Fill small gaps
        mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)  # Remove noise


        # Extract silhouette
        final = cv2.bitwise_and(img, img, mask=mask)

        # Compute FPS
        curr_time = time.time()
        fps = round(1 / (curr_time - prev_time), 2)
        prev_time = curr_time
        print(f"FPS: {fps}")

        # Display results
        cv2_imshow(mask)
        cv2_imshow(final)

        # Press 'Esc' to exit
        if cv2.waitKey(10) == 27:
            break

    video.release()
    cv2.destroyAllWindows()
```
